```python
import pandas as pd
from pathlib import Path
```

```python
tracks = pd.read_csv(Path("./clean_tracks.csv"))
tracks2 = pd.read_csv(Path("./top5000_cleaned.csv"))
```

```python
tracks_df1 = pd.DataFrame(tracks)
tracks_df2 = pd.DataFrame(tracks2)
```

```python
df1 = tracks_df1.drop(columns='release_date')
df_sort = df1.sort_values('popularity', ascending=False)
df_sort.reset_index(drop=True)
df_sort.head()
```

```python
df_sort1 = tracks_df2.sort_values('popularity', ascending=False)
```

```python
df2 = df_sort1.reset_index(drop=True)
```

```python
frames = [df_sort, df2]
```

```python
df_main = pd.concat(frames)
```

```python
df_main.duplicated().sum()
```

```python
df_clean = df_main.drop_duplicates()
```

```python
df_clean1 = df_clean.drop_duplicates(subset=['track_id'])
df_clean2 = df_clean.drop_duplicates(subset=['track_name'])
```
