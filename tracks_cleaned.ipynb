```python
import pandas as pd
from pathlib import Path
```

```python
tracks = pd.read_csv(Path("tracks.csv"))
df = pd.DataFrame(tracks)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
clean_df = df.dropna()
```

```python
clean_df.duplicated().sum()
```

```python
#tops = clean_df.sort_values(by=['popularity'], ascending=False, inplace=False)
#tops.head()
```

```python
three_th = clean_df.sample(n=5000)
tracks_df = pd.DataFrame(three_th)
tracks_df.info()
```

```python
tracks_df['release_date'] = pd.to_datetime(tracks_df['release_date'], format="%Y/%m/%d")
```

```python
tracks_df.set_index('release_date', inplace=True)
```

```python
tracks_df.sample(20)
```

```python
tracks_df['duration'] = round(tracks_df['duration_ms']/1000)
```

```python
track_clean = tracks_df.drop(columns='duration_ms')
```

```python
track_clean['artists'] = track_clean['artists'].str.replace(r'[][\'\,]', '', regex=True)
track_clean['id_artists'] = track_clean['id_artists'].str.replace(r'[][\'\,]', '', regex=True)
```

```python
track_clean.head()
```

```python
track_clean = track_clean[['artists', 'id', 'name', 'acousticness', 'danceability', 'energy',
                          'instrumentalness', 'key', 'liveness', 'loudness', 'mode', 'speechiness', 'tempo',
                          'time_signature', 'valence', 'popularity', 'duration']]
```

```python
track_clean.columns = ['artist_name', 'track_id', 'track_name','acousticness', 'danceability', 'energy',
                          'instrumentalness', 'key', 'liveness', 'loudness', 'mode', 'speechiness', 'tempo',
                          'time_signature', 'valence', 'popularity', 'duration']
```

```python
track_clean.head()
```

```python
track_clean.to_csv('clean_tracks.csv', index=True)
```
