```python
import pandas as pd
import os
```

```python
# Read the Spotify data file and store it in a pandas DataFrame
spotify_data = 'SpotifySongPopularity.csv'
spotify_df = pd.read_csv(spotify_data)
```

```python
spotify_df.head()
```

```python
spotify_df.info()
```

```python
spotify_df.isnull().sum()
```

```python
cleaned_spotify_df = spotify_df.dropna()
```

```python
cleaned_spotify_df.duplicated().sum()
```

```python
cleaned_spotify_df.info()
```

```python
cleaned_spotify_df['duration'] = round(cleaned_spotify_df['duration_ms']/1000)
```

```python
cleaned_spotify_df.head()
```

```python
clean_spotify_df = cleaned_spotify_df.drop(columns={'duration_ms'})
clean_spotify_df.head()
```

```python
top_spotify_songs = clean_spotify_df.sort_values(by=['popularity'], ascending=False, inplace=False)
top_spotify_songs.head(10)
```

```python
top_3000 = top_spotify_songs[0:3000]
top_3000_df = pd.DataFrame(top_3000)
top_3000_df.info()
```

```python
top_3000_df.head()
```

```python
top_3000.to_csv("top3000_cleaned.csv", index=False)
```
