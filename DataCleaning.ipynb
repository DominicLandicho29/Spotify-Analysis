```python
#importing libraries 
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("Spotify_final_dataset.csv")
df
```

```python
# Check the data types
df.info()
```

```python
# Check the null values
df.isnull().sum()
```

```python
# Drop the null values
df=df.dropna()
```

```python
# Convert the Peak Position (xTimes) column to integer type and remove unwanted symbols
df['Peak Position (xTimes)'] = df['Peak Position (xTimes)'].str.replace('(', '')
df['Peak Position (xTimes)'] = df['Peak Position (xTimes)'].str.replace('x', '')
df['Peak Position (xTimes)'] = df['Peak Position (xTimes)'].str.replace(')', '')
df["Peak Position (xTimes)"]=df["Peak Position (xTimes)"].astype(int)
```

```python
df.info()
```

```python
# Convert the Top 10 (xTimes) Column to Integer 
df["Top 10 (xTimes)"]=df["Top 10 (xTimes)"].astype(int)
```

```python
df.head()
```

```python
#Rename the Top 10 (xTimes) column to not include a number
df = df.rename(columns={'Top 10 (xTimes)':'Top Ten (xTimes)'})
```

```python
df.head()
```

```python
# Export the dataframe into a csv
df.to_csv('Spotify_Final_dataset_cleaned.csv', index=False)
```
